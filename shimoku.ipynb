## How have genre trends evolved over the years?

```python
import datetime as dt
```

```python
genres_df=all_titles[['title', 'release_year', 'genres']]
```

```python
genres_df=genres_df.sort_values(by='release_year')
```

```python
genres_df.tail()
```

```python
genre_counts_by_year = {}
for index, row in genres_df.iterrows():
    year = row['release_year']
    genres = row['genres']
    if year not in genre_counts_by_year:
        genre_counts_by_year[year] = {genre: 0 for genre in all_genres}
    for genre in genres:
        if genre in genre_counts_by_year[year]:
            genre_counts_by_year[year][genre] += 1
result = [{'date': dt.date(year, 1, 1), **genre_counts}
          for year, genre_counts in genre_counts_by_year.items()]
```

```python
result
```

```python
genres_df_1900=genres_df[genres_df['release_year']<2000]
```

```python
genre_counts_by_year = {}
for index, row in genres_df_1900.iterrows():
    year = row['release_year']
    genres = row['genres']
    if year not in genre_counts_by_year:
        genre_counts_by_year[year] = {genre: 0 for genre in all_genres}
    for genre in genres:
        if genre in genre_counts_by_year[year]:
            genre_counts_by_year[year][genre] += 1
result_1900 = [{'date': dt.date(year, 1, 1), **genre_counts}
          for year, genre_counts in genre_counts_by_year.items()]
print(result_1900)
```

```python
result_1900
```

```python
genres_df_2000=genres_df[genres_df['release_year']>=2000]
```

```python
# Initialize an empty dictionary to store genre counts by year
genre_counts_by_year = {}

# Iterate over each row in the DataFrame
for index, row in genres_df_2000.iterrows():
    year = row['release_year']
    genres = row['genres']

    # Create a new entry for the year if it doesn't exist in the dictionary
    if year not in genre_counts_by_year:
        genre_counts_by_year[year] = {genre: 0 for genre in all_genres}

    # Update genre counts for the year
    for genre in genres:
        if genre in genre_counts_by_year[year]:
            genre_counts_by_year[year][genre] += 1

# Convert the dictionary to the desired format (list of dictionaries)
result_2000 = [{'date': dt.date(year, 1, 1), **genre_counts}
          for year, genre_counts in genre_counts_by_year.items()]

# Display the result
print(result_2000)
```

```python
platforms
```

```python
ages=platforms_df_age['age_certification'].unique()
```

```python
ages
```

```python
pd.cut(platforms_df_age['runtime'],10)
```

```python
result_age = []
for platform in platforms:
    platform_dict = {'platform': platform}
    for age_cert in ages:
        count = platforms_df_age[(platforms_df_age['streaming'] == platform) & (platforms_df_age['age_certification'] == age_cert)].shape[0]
        platform_dict[age_cert] = count
    result_age.append(platform_dict)
```
